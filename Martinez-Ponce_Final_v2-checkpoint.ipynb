```python
response=requests.get("https://finance.yahoo.com/quote/DIS/financials?p=DIS")
web_page=response.text
soup = BeautifulSoup(web_page, "html.parser")
features = soup.find_all('div', class_='D(tbr)')
headers = []
temp_list = []
label_list = []
final = []
index = 0
for item in features[0].find_all('div', class_='D(ib)'):
    headers.append(item.text)
while index <= len(features)-1:
    #filter for each line of the statement
    temp = features[index].find_all('div', class_='D(tbc)')
    for line in temp:
        #each item adding to a temporary list
        temp_list.append(line.text)
    #temp_list added to final list
    final.append(temp_list)
    #clear temp_list
    temp_list = []
    index+=1
DisIncome=pd.DataFrame(final[1:])
DisIncome.columns=headers

DisIncome['ttm']=DisIncome['ttm'].str.replace(',','')
DisIncome['9/30/2020']=DisIncome['9/30/2020'].str.replace(',','')
DisIncome['9/30/2019']=DisIncome['9/30/2019'].str.replace(',','')
DisIncome['9/30/2018']=DisIncome['9/30/2018'].str.replace(',','')
DisIncome['9/30/2017']=DisIncome['9/30/2017'].str.replace(',','')
DisIncome=DisIncome.replace('-','0')
DisIncome['ttm']=DisIncome['ttm'].astype(int)
DisIncome['9/30/2020']=DisIncome['9/30/2020'].astype(float)
DisIncome['9/30/2019']=DisIncome['9/30/2019'].astype(float)
DisIncome['9/30/2018']=DisIncome['9/30/2018'].astype(float)
DisIncome['9/30/2017']=DisIncome['9/30/2017'].astype(float)
#Revenue
DisRev=DisIncome.loc[[0]]
DisRev=DisRev.drop(columns='ttm')
DisRev=DisRev.drop(columns='Breakdown')
DisRev=DisRev.transpose()
#RevenuePlot
DisRev.plot.bar()
plt.title('Total Revenue')
plt.ylabel('Total Revenue(in Millions)')
plt.xlabel('Date')
plt.title('Total Revenue')
plt.ylabel('Total Revenue(in Millions)')
plt.xlabel('Date')
plt.show()
```
