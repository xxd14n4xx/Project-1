```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("df.csv")
df.head()
```

```python
df=df.drop(columns='Unnamed: 0')
```

```python
a=df.groupby('Sport')['Pay'].mean()
a
```

```python
a.plot.bar()
```

```python
df['Pay']=df['Pay']/1000000
df
```

```python
a=df.groupby('Sport')['Pay'].mean()
a.plot.bar()
```

```python
#Plot 1
a.plot.bar()
plt.title('Average Pay by Sport', fontdict={'fontname':'Arial','fontsize':'15','color':'navy'}, loc='right')
plt.ylabel('Pay in ($Million)', fontdict={'fontname':'Arial', 'fontsize':10})
plt.xlabel('Sport', fontdict={'fontname':'Arial', 'fontsize':10})
plt.axhline(y=27.5, color='red', label='Average Pay Line', linestyle='--', linewidth=2)
plt.text(8.4,28,'27.5 M', color='red')
plt.legend()
plt.show()
#Here I can see that Boxing players on average get paid the most.
#The least paid are Baseball and Football players (On average)
```

```python
df[['Pay']].mean()
```

```python
b=df.groupby('Sport')[['Endorsements']].mean()
b=b['Endorsements']/1000000
b
```

```python
#Plot 2
b.plot.bar(color='black')
plt.ylabel('Pay in ($Million)', fontdict={'fontname':'Arial', 'fontsize':10})
plt.xlabel('Sport', fontdict={'fontname':'Arial', 'fontsize':10})
plt.show
```

```python
#Plot 3
a.plot.bar(color='blue')
b.plot.bar(color='yellow')
plt.ylabel('Pay in ($Million)', fontdict={'fontname':'Arial', 'fontsize':10})
plt.xlabel('Sport', fontdict={'fontname':'Arial', 'fontsize':10})
plt.legend()
plt.show()
```

```python
#Plot 4
rows=df.groupby(['Sport']).size()
rows
```

```python
Base=27
Bask=18
Box=4
Crick=1
Foot=17
Golf=5
Raci=6
Socc=15
Tenn=6
Trac=1
plt.pie([Base,Bask,Box,Crick,Foot,Golf,Raci,Socc,Tenn,Trac],
        labels=['Baseball','Basketball','Boxing','Cricket','Football','Golf','Racing','Soccer','Tennis','Track'], autopct='%.2f %%')
plt.show()
```

```python
#Plot 5
plt.scatter(df['Rank'],(df['Pay']/1000000))
plt.ylabel('Pay in ($Million)', fontdict={'fontname':'Arial', 'fontsize':15})
plt.xlabel('Rank', fontdict={'fontname':'Arial', 'fontsize':15})
#If your rank is more than 20, you are more likely to get paid less than 30 million. 
#Top ranked athletes get paid more
```

```python
#Plot 6
plt.hist(df['Height'],color='green')
plt.ylabel('Frequency', fontdict={'fontname':'Arial', 'fontsize':15})
plt.xlabel('Height', fontdict={'fontname':'Arial', 'fontsize':15})
plt.show()
#Mostly every athlete is around 190 cm
```
