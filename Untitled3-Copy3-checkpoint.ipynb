```python
athNew=(ath[['Name','Rank','Total Pay','Salary/Winnings','Endorsements']])
athNew
#Here I  created a new DataFrame
```

```python
athNew.dtypes
```

```python
athNew=athNew.rename(columns={'Total Pay': 'Pay','Salary/Winnings':'Salary'})
athNew.head()
##Rename Columns
```

```python
athNew['Pay','Salary','Endorsements']=athNew['Pay','Salary','Endorsements'].astype(int)
athNew.dtypes
```

```python
athNew['Pay']=athNew['Pay'].str.replace('$','')
athNew.head()
```

```python
athNew['Pay']=athNew['Pay'].str.replace(',','')
athNew.head()
```

```python
athNew['Salary']=athNew['Salary'].str.replace('$','')
athNew.head()
```

```python
athNew['Salary']=athNew['Salary'].str.replace(',','')
athNew.head()
```

```python
athNew['Endorsements']=athNew['Endorsements'].str.replace('$','')
athNew.head()
```

```python
athNew['Endorsements']=athNew['Endorsements'].str.replace(',','')
athNew.head()
```

```python
athNew['Pay']=athNew['Pay'].astype(int)
athNew.dtypes
```

```python
athNew['Salary']=athNew['Salary'].astype(int)
athNew.dtypes
```

```python
athNew['Endorsements']=athNew['Endorsements'].astype(int)
athNew.dtypes
```

```python
athNew.corr()
#Pay and Salary are the varieables that have the highest correlation
```
